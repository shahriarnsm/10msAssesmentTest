```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("case_study.xlsx","Dataset - Excel Test 1  Basic 1" )
df.head()
```

```python
df.drop(labels=["Vlookup Table", "Unnamed: 1", "Unnamed: 2"], axis = 1).head()
```

```python
grouped = df.groupby("Category")["Sales Amount", "Net Revenue"].sum()
```

```python
##Sale ammount & net revenue comparisn
grouped.plot.bar(figsize=(20, 6))
```

```python
grouped = df.groupby('Category').sum()
```

```python
grouped = df.groupby('Attribution')
costs = grouped[['C_Cost', 'A_cost']].sum()
```

```python
grouped.plot(kind='bar', figsize=(12, 6))
```

```python
# grouped.plot.bar(figsize=(18, 6))
# grouped = df.groupby("Category")["Sales Amount", "Net Revenue"].sum()
costs.plot(kind='bar', figsize=(16, 6))
plt.xlabel('Category')
plt.ylabel('Cost')
plt.title('Campaign and Affiliate Cost by Product Category')
plt.legend(['Campaign Cost', 'Affiliate Cost'])
plt.show()
```

```python
# Group the data by category and calculate the sum of the net revenue
grouped = df.groupby("Attribution").sum()["Net Revenue"]

# Create a bar graph to visualize the comparison
grouped.plot(kind="bar", figsize=(20, 6))
plt.title("Compare the Net Revenue of all 4 Category of Sales (Campaign, Affiliate, Organic and Direct).")
plt.xlabel("Category")
plt.ylabel("Net Revenue")
plt.show()
```
