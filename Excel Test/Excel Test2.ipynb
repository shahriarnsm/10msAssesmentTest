```python
## Handaling missing Data or error Data
```

```python
# Count the number of NaN values in each column
nan_count = df.isna().sum()

# Get the column names that have NaN values
nan_cols = nan_count[nan_count > 0].index

# Print the result
print("Columns with NaN values:")
for col in nan_cols:
    print("-", col, ":", nan_count[col])
```

```python
new_df = df.interpolate()
new_df
```

```python
# Check the number of NaN values in each column
nan_count = new_df.isna().sum()

# Get the column names that have NaN values
nan_cols = nan_count[nan_count > 0].index

# Print the result
print("Columns with NaN values:")
for col in nan_cols:
    print("-", col, ":", nan_count[col])
```

```python
# create the pivot table with average values of reach, engagement, views, and clicks(with NaN values)
pivot_table = df.pivot_table(index="Content type",values=["Reach", "Engagement", "View", "Clicks"], aggfunc="mean")
# display the pivot table
print(pivot_table)
```

```python
# create the pivot table after fill the NaN value with average values of reach, engagement, views, and clicks
pivot_table = new_df.pivot_table(index="Content type",values=["Reach", "Engagement", "View", "Clicks"], aggfunc="mean")
# display the pivot table
pivot_table
```

```python
# filter data for only "REGULAR POSTERS"
df = df[df['Content type'] == 'REGULAR POSTER']

# calculate the moving average with a 7-day window
df['Clicks - Moving Average'] = df['Clicks'].rolling(window=7).mean()

# print the result
print(df)
```
